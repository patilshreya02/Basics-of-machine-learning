```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.regplot(x="LotArea",y="SalePrice",data=df) #bivariate (scatter with reg line)   outliers where lotarea is high and sale is low or vice versa
```

```python
df["LotArea"].quantile([0.5,0.95,0.99]) #outlietrs  #0.5 is median and high val is 37567.64
```

one way to remove outlier is to use quantile find out quantile(median val)and if u feel after certain val if certain val high then filter it7yg  

```python
df_sub = df.loc[df["LotArea"]< df["LotArea"].quantile(0.95)] #remove outliers considering data till 0.95 so there is variations as lot area  increases sale also increases

sns.regplot(x="LotArea",y="SalePrice",data=df_sub)
```

```python
sf_cols=[col_name for col_name in df.columns if "SF" in col_name]
fig,axs = plt.subplots(nrows=3,ncols=3,figsize=(10,10))
sns.regplot(x="LotArea",y="SalePrice",data=df)
```

```python
for i in range(0,len(sf_cols)):
    rows=i//3                  #i is 0. 0//3 ,1//3, 2//3 so on
    cols=i%3
    print(rows,cols)
```

```python
ax=axs[rows,cols]                                            # filter the column which matches sf 
plot = sns.regplot(x=sf_cols[i],y="SalePrice",data=df,ax=ax)
```
